Analysis
====================

In this mini project, we generated parameter ICPC, ML and sequences with specific sampled position. Applying the motif finder to these sequences with different combination of parameters, we generated 70 PWM and recorded the starting sites as well.  In order to assess the motif finder, we compared the experimental finding with theoretical results by several methods.

Relative Entropy
-------------------
The first method is to evaluate the relative entropy between experimental PWM and theoretical PWM with respect to changes in the different parameters, ICPC, motif length, sequence count. 

```python
import analysis
analysis.rel_icpc()
```

In this figure, we can observe relative entropy and standard deviation decreasing with an increase in ICPC, which means the experimental PWMs are more reliable (shorter error bar) and more similar to the theoretical one. Thus, in the ICPC range from 1.0-2.0, the larger ICPC is, the better motif finding program is able to recover the motif.

```python
analysis.rel_ml()
```

In this figure, we can observe that changes in motif length does not seem to affect our ability to recover the motif.

```python
analysis.rel_sc()
```

As can be seen, changes in sequence count did not really result in any change in the relative entropy.

Sites Overlap
--------------

```python
analysis.sites_icpc()
```

As observed, an increased in ICPC led to us being able to recover the motif better. As can be seen, when the ICPC is too low, we are nearly unable to recover any of the motifs.

```python
analysis.sites_ml()
```

As can be seen here, changes in motif length does not really affect our ability to recover the full motif. We are still mostly able to recover the whole motif.

```python
analysis.sites_sc()
```

Changes in number of sequences seem to only slightly affect how well we are able to recover our motif.

Time taken
--------------
Our third measure is to plot the time taken against changes in the various parameters.

```python
analysis.time_icpc()
```

With an increase in the information content, we can observe that the time taken to find each motif is going down. This is expected as with a greater ICPC, the algorithm will be more likely to alight on a satisfactory motif sooner. As our algorithm exits as soon as the current motif's score crosses a certain threshold, the higher information content allows us to return earlier.

```python
analysis.time_ml()
```

As can be seen, the longer the motif length, the longer the time taken for the algorithm to complete. This is fits our expectation as well, since the algorithm now needs to check one more spot in each motif that it is generating.

```python
analysis.time_sc()
```

As can be quite clearly seeen here, the time taken increases as the number of sequences we are looking at increase. Since the algorithm is random, the more sequences there are, the harder it is for it to randomly find a shared motif between all the sequences.

Conclusion
-----------

Increasing the informational content seems to have the best result. However, a finer resolution is required in order for us to draw better conclusions.
